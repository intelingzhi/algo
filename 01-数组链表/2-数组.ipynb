## 数组双指针

### 26. 删除有序数组中的重复项

https://leetcode.cn/problems/remove-duplicates-from-sorted-array/description/

快慢指针

```python
class Solution(object):
    def removeDuplicates(self, nums):
        if len(nums) == 1:
            return 1
        # 定义快慢指针
        slow = 0
        fast = 0

        while fast < len(nums):
            if nums[fast] != nums[slow]:
                slow += 1
                # 让nums[0]到nums[slow]没有重复
                nums[slow] = nums[fast]

            fast += 1
```

### 83. 删除排序链表中的重复元素

https://leetcode.cn/problems/remove-duplicates-from-sorted-list/description/

同样是使用快慢指针，原理和上面一样，但是要注意最后一句要断开slow后面的东西。

```python
class Solution(object):
    def deleteDuplicates(self, head):
        # 边界判断非常重要。拿到一个输入，要想着他是不是会害我。
        if head is None:
            return None

        slow = head
        fast = head

        while fast is not None:
            if slow.val != fast.val:
                slow.next = fast
                slow = slow.next

            fast = fast.next
        # 断开与后面元素的连接（这一步很重要）
        slow.next = None
        return head
```

### 27. 移除元素

https://leetcode.cn/problems/remove-element/description/

同样使用快慢指针技巧。

```python
class Solution(object):
    def removeElement(self, nums, val):
        if len(nums) == 0:
            return 0
        
        slow = 0
        fast = 0

        while fast < len(nums):
            if nums[fast] != val:
                nums[slow] = nums[fast]
                slow += 1
            
            fast += 1

        return slow
```

### 283. 移动零

https://leetcode.cn/problems/move-zeroes/description/

```python
class Solution(object):
    def moveZeroes(self, nums):
        if len(nums) == 1:
            return nums

        # 去除0
        pos = self.removeElement(nums, 0)

        # 后面统一置为0
        for i in range(pos, len(nums)):
            nums[i] = 0
        


    def removeElement(self, nums, val):
        slow = 0
        fast = 0

        while fast < len(nums):
            if nums[fast] != 0:
                nums[slow] = nums[fast]
                slow += 1

            fast += 1

        return slow
```

## 滑动窗口
滑动窗口算法技巧主要用来解决子数组问题，比如让你寻找符合某个条件的最长/最短子数组。

### 76. 最小覆盖子串

https://leetcode.cn/problems/minimum-window-substring/description/

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # 记录窗口的字符
        window ={}
        # 需要凑齐的字符
        need = {}

        # 统计字符串中每个字符出现的次数
        # 若t="aabbc"，则 need = {'a': 2, 'b': 1, 'c': 2}
        for c in t:
            need[c] = need.get(c, 0) + 1

        # 区间 [left, right) 是左闭右开的，所以初始情况下窗口没有包含任何元素
        left = 0
        right = 0
        valid = 0

        # 记录最小覆盖串的起始索引和长度
        start = 0
        length = float('inf') # 正无穷大

        while right < len(s):
            # c 是移入窗口的字符
            c = s[right]
            right += 1

            # 窗口数据更新。如果这个c是要求内的，就加到window里面
            if c in need:
                window[c] = window.get(c, 0) + 1
                if window[c] == need[c]: # 如果其中一个字母完全匹配，则完成当前字母的任务
                    valid += 1

            # 判断收缩
            while valid == len(need): # 满足条件就收缩，否则直接跳过
                # 检查这次的串是否小于已知的最小的串
                if right - left < length:
                    start = left
                    length = right - left

                # 移除头字母
                d = s[left]
                left += 1

                if d in need:
                    # 如果当前移除的字母破坏了最小元素原则，则说明砍多了
                    if window[d] == need[d]:
                        valid -=1
                    window[d] -= 1


        return "" if length == float('inf') else s[start: start+length]
```

### 567. 字符串的排列

https://leetcode.cn/problems/permutation-in-string/description/

```python
class Solution:
    def checkInclusion(self, t: str, s: str) -> bool:
        window ={}
        need = {}
        for c in t:
            need[c] = need.get(c, 0) + 1

            
        left = 0
        right = 0
        valid = 0

        # 移入窗口 (right来负责)
        while right < len(s):
            c = s[right]
            right += 1

            # 光是加进去不行，还得审核一下是否是内部人员
            if c in need:
                window[c] = window.get(c, 0) + 1
                # 再检查一下内部人员的人数齐了没
                if window[c] == need[c]:
                    valid += 1


            # 移出窗口 (left来负责)
            # 也就是窗口值达到匹配串以上，就开始查
            # 这里的while可以换成if，因为这里本质上是一个定长窗口
            while right - left >= len(t):
                if valid == len(need):
                    return True
                
                d = s[left]
                left += 1

                if d in need:
                    if window[d] == need[d]:
                        valid -= 1
                    window[d] -= 1

        return False
                
```
