# 链表算法

### 21. 合并两个有序链表

https://leetcode.cn/problems/merge-two-sorted-lists/description/

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

"""
:type list1: Optional[ListNode]
:type list2: Optional[ListNode]
:rtype: Optional[ListNode]
"""

class Solution(object):
    def mergeTwoLists(self, list1, list2):
        # 这个-1是随便选的，为了不干扰正常数据，选的-1，也可以选择-2 -3 -4等等
        dummy = ListNode(-1)
        p = dummy
        p1 = list1
        p2 = list2

        # 开始拉链
        while p1 is not None and p2 is not None:
            if p1.val > p2.val:
                p.next = p2
                p2 = p2.next

            else:
                p.next = p1
                p1 = p1.next

            p = p.next

        if p1 is not None:
            p.next = p1

        if p2 is not None:
            p.next = p2

        return dummy.next
```

### 86. 分隔链表

https://leetcode.cn/problems/partition-list/

```python

"""
:type head: Optional[ListNode]
:type x: int
:rtype: Optional[ListNode]
"""

class Solution(object):
    def partition(self, head, x):
        dummy1 = ListNode(-1)  # 小于x的链表
        dummy2 = ListNode(-1)  # 大于等于x的链表
        p1, p2 = dummy1, dummy2

        p = head

        while p:
            if p.val >= x:
                p2.next = p
                p2 = p2.next
            else:
                p1.next = p
                p1 = p1.next

            # 不能直接next
            # p = p.next

            # 需要断开后面的节点
            # p.next = None

            # 断开之前先保存
            tmp = p.next
            p.next = None
            p = tmp

        p1.next = dummy2.next

        return dummy1.next
        
```

### 23. 合并 K 个升序链表

https://leetcode.cn/problems/merge-k-sorted-lists/description/

```python
import heapq

# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next



class Solution(object):
    def mergeKLists(self, lists):
        # not lists 等价于 len(lists) == 0
        if not lists:
            return None
        
        dummy = ListNode(-1)
        p = dummy

        priorityQueue = []

        # 将所以链表的头节点加入最小堆
        # 添加索引 i 作为第二个元素
        # 当两个节点的值 val 相同时，Python 会尝试比较元组中的下一个元素，也就是i
        for i, head in enumerate(lists):
            if head is not None:
                heapq.heappush(priorityQueue, (head.val, i, head))

        """
        堆内结构
                 (1,0)
                /     \
             (1,1)   (2,2)
        """

        while priorityQueue:
            # 获取最小节点加入最小堆
            val, i, node = heapq.heappop(priorityQueue)
            p.next = node
            if node.next is not None:
                heapq.heappush(priorityQueue, (node.next.val, i, node.next))

            p = p.next

        return dummy.next
```
