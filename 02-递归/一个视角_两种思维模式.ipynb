```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

# 把递归看作是一棵树

```python
# 全排列问题

class Solution:
    def __init__(self):
        self.res = []

    # 主函数
    def permute(self, nums):
        # 记录路径
        track = []

        # 记录是否使用，初始状态下都是false
        used = [False] * len(nums)

        self.backtrack(nums, track, used)
        return self.res
    
    def backtrack(self, nums, track, used):
        # 先写叶子节点的逻辑
        if len(track) == len(used):
            self.res.append(track.copy())

        # 开始写树节点的逻辑，相当于多叉树遍历
        for i in range(len(nums)):
            # 排除不合法
            if used[i]:
                continue

            track.append(nums[i])
            used[i] = True

            # 进入下一层决策树
            self.backtrack(nums, track, used)

            # 取消选择
            track.pop()
            used[i] = False

```

# 两种思维

```python
# 分解问题的思维模式
# 如果你想用「分解问题」的思维模式来写递归算法，那么这个递归函数一定要有一个清晰的定义，说明这个函数参数的含义是什么，返回什么结果。

# 用分解思路求fib
def fib(n):
    if n < 2:
        return n

    # 子问题
    fa = fib(n-1)
    fb = fib(n-2)

    # 由子问题的解，得到原问题的解
    return fa + fb



# 用分解思路求二叉树最大深度
# 分解思路：计算左右子树的各自的最大深度，然后取最大值
# 明确maxDepth函数的输入输出
#    输入：（子）二叉树的头节点
#    输出：这棵树的最大深度
class Solution:
    def maxDepth(self, root: TreeNode) -> int:
        if root is None:
            return 0
        
        leftMax = self.maxDepth(root.left)
        rightMax = self.maxDepth(root.right)

        return 1 + max(leftMax, rightMax)
    
# 用遍历思路求二叉树最大深度
# 如果你想用「遍历」的思维模式来写递归算法，那么你需要一个无返回值的遍历函数，在遍历的过程中收集结果。
# 这里的一个特点时，只有走到叶子节点才算得到一个结果。

class Solution2:
    def __init__(self):
        self.depth = 0
        self.maxHeight = 0

    def maxDepth(self, root: TreeNode) -> int:
        self.traverse(root)
        return self.res
    
    def traverse(self, root: TreeNode):
        if root is None:
            return
        
        # 前序位置：进入节点之前+1
        self.depth += 1

        # 遍历到叶子节点
        if root.left is None and root.right is None:
            self.maxHeight = max(self.maxHeight, self.depth)

        self.traverse(root.left)
        self.traverse(root.right)

        # 后序遍历位置
        self.depth -= 1
```
