```python
from typing import List, Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# 21. 合并两个有序链表

# 1、读取两个指针
# 2、定义返回结构（自己造链表，带虚拟头节点）
# 3、判断结束条件
# 4、判断拉链
# 5、干活
# 6、返回

class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        # 定义双指针
        p1 = l1
        p2 = l2

        # 定义返回结构
        dummy = ListNode(-1)
        p = dummy

        # 判断条件是只要有一个走完就结束
        if p1 is not None and p2 is not None:
            # 开始拉链
            if p1.val > p2.val:
                p.next = p2
                p2 = p2.next
            else:
                p.next = p1
                p1 = p1.next

            p = p.next

        if p1 is not None:
            p.next = p1
        if p2 is not None:
            p.next = p2

        return dummy.next
```

```python
# 86. 分隔链表

# 1、定义并读取两个链表（自己创建的链表，需要带虚拟节点）
# 2、判断主链表，按照要求把节点塞进两个链表里
# 3、连接两个链表
# 4、返回

class Solution:
    def partition(self, head: ListNode, x: int) -> ListNode:
        
        dummy1 = ListNode(-1)
        dummy2 = ListNode(-1)

        p1, p2 = dummy1, dummy2

        p = head

        while p is not None:
            if p.val >= x:
                p2.next = p
                p2 = p2.next
            else:
                p1.next = p
                p1 = p1.next

            # 考虑一种情况
            # 如果不处理链表的 next 的话，可以，但是最后一个就不会自动处理了。
            # 所以还是自己处理吧。
            tmp = p.next    # 取值
            p.next = None   # 断联
            p = tmp

        p1.next = dummy2.next
        return dummy1.next
```

```python
# 23. 合并 k 个升序链表

# 1、创建链表，带隐藏节点
# 2、创建优先级队列
# 3、将头节点加入最小堆
# 4、开始重放，每次从堆中取最小值

import heapq

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    # 重载比较运算符，方便将 ListNode 加入最小堆
    def __lt__(self, other):
        return self.val < other.val
    

class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if lists is None:
           return None
        
        dummy = ListNode(-1)
        p = dummy

        # 最小堆
        pq = []

        # 加入最小堆
        for idx, head in enumerate(lists):
            if head is not None:
                # head.val 用于比较。如果一样，则比较 idx
                # idx肯定是不一样的，这里的head 主要是信息存储
                heapq.heappush(pq, (head.val, idx, head))
        
        # 重放
        while pq:
            val, idx, node = heapq.heappop(pq)
            p.next = node

            # 头节点已用，寻找接替者
            if node.next is not None:
                heapq.heappush(pq, (node.next.val, idx, node.next))

            p = p.next

        return dummy.next
```
